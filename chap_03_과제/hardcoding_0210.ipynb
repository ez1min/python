로또 생성기
======================
1) 6가지 숫자가 선택되어야 해요.
2) 중복되면 안되요.
3) 오름차순으로 정리해 주세요.
======================
4) 10만회 돌리고
각각 번호가 나온 것을 카운팅하기
5) 각 번호를 퍼센트 내기(소숫점 두자리)
6) 제일 높게 나온 확률 순서대로 번호 3개 추천해주기

# 

```python
import random as r

box_list = []        # 모든 리스트 저장
res_list = []        # 몇 번 나왔는지 카운트 리스트
number_list = []     # 1~45까지 숫자 리스트

# 100번 리스트 생성
for _ in range(100):
    box = sorted(r.sample(range(1, 46), 6))  # 1~45 중 오름차순으로 6개 숫자 뽑기
    box_list.extend(box)                    # 모든 숫자 저장
    print(box)                              # 생성된 리스트 출력

# 1~45까지의 숫자 등장 횟수, 확률 계산
for i in range(1, 46):
    count = box_list.count(i)               # i의 등장 횟수
    percent = count * 100 / len(box_list)   # 등장 확률 (%)

    print("{%d}가 {%d}번 나옴, 확률은 %.2f%%" % (i, count, percent))  # 소숫점 두 번째 자리까지 계산

    res_list.append(count)                  # 카운트 결과 저장
    number_list.append(i)                   # 숫자 저장

# 등장 횟수와 숫자를 튜플로 묶기
tuple_plus = list(zip(number_list, res_list))
print("숫자와 등장 횟수:", tuple_plus)

# 가장 많이 등장한 숫자 찾기
max_value = max(tuple_plus, key=lambda x: x[1])
print("가장 많이 나온 숫자:", max_value[0], "등장 횟수:", max_value[1])

# 두 번째로 많이 나온 숫자 찾기
tuple_plus = sorted(tuple_plus, key=lambda x: x[1], reverse=True)  # 등장 횟수 기준 내림차순 정렬
second_largest_value = tuple_plus[1][1]                             # 2번째로 많이 나온 숫자는 등장 횟수를 내림차순으로 정렬 하였으니 두 번째 요소임  
find_sec_tuple = [x for x in tuple_plus if x[1] == second_largest_value]  #tuple plus 리스트의 각 튜플을 순회하고, 두 번째 요소가 second_largest_value와 같은 경우에 필터링 하고 새 리스트에 저장 함
first_elements = [x[0] for x in find_sec_tuple]   #find_sce_tuple 리스트의 각 튜플을 순회하고, 각 튜플의 첫번째 요소를 추출 하고 저장 한다

#결과 출력
print("2번째로 많이 나온 숫자:",first_elements, "등장 횟수:",second_largest_value)


tuple_plus = sorted(tuple_plus, key=lambda x: x[1], reverse=True) #세 번째로 많이 나온 숫자 찾기
# 등장 횟수 기준 내림차순 정렬

third_largest_value = tuple_plus[2][1]  # 3번째로 많이 나온 숫자는 내림차순 정렬된 리스트의 세 번째 요소

find_third_tuple = [x for x in tuple_plus if x[1] == third_largest_value] # 등장 횟수가 third_largest_value와 같은 튜플 필터링

third_elements = [x[0] for x in find_third_tuple]    # 해당 튜플의 첫 번째 요소(숫자) 추출

recommend_numbers = set()                   # 중복 제거를 위한 set 사용
recommend_numbers.add(max_value[0])         # 가장 많이 나온 숫자 추가
recommend_numbers.update(first_elements)    # 두 번째로 많이 나온 숫자 추가
recommend_numbers.update(third_elements)    
 
# 결과 출력
print("3번째로 많이 나온 숫자:", third_elements, "등장 횟수:", third_largest_value)


print("추천 숫자:", max_value[0], first_elements, third_elements)
```

## 잔돈계산 

```python
money = []                         #투입 금액을 리스트에 넣기위한 변수
a = []                             #구매할 물건의 가격을 넣기위한 변수

print("돈을 넣어주세요")
t = int(input())   
money.append(t)
print("투입?: {%d}원" %t)

m = [1000, 2060, 3320, 4500]
print(m)
print("어떤 것을 구매하나요?: ")

c = int(input())                   # 구매할 물건의 가격 입력
a.append(c)                        # 리스트 a에 입력 받은 c값을 넣어줌 

# 거스름돈 계산
res = int(money[0] - a[0])      

print("거스름돈: {%d}원" %res)

# 1000원 단위 계산
thousand = res // 1000             #몫 연산을 이용해서 장(개)단위로 거슬러 줌
remainder = res % 1000             #1000원으로 거슬러 주고 남은돈은 나머지 연산을 이용해서 변수에 넣음

# 500원 단위 계산
five_hundred = remainder // 500   
remainder = remainder % 500        

# 100원 단위 계산
hundred = remainder // 100
remainder = remainder % 100        

# 10원 단위 계산
ten = remainder // 10

# 거스름돈 출력
if thousand != 0:                                #거슬러 줄 1000원짜리가 있다면(0이 아님) 출력 하고 , 없다면 밑줄로 이동         
    print("거스름돈: 1000원: {%d}장" %thousand)   #1000원 짜리로 거슬러 줌
   
if five_hundred != 0:                             
    print("거스름돈: 500원: {%d}개" %five_hundred)

if hundred != 0:
    print("거스름돈: 100원: {%d}개" %hundred)

if ten != 0:
    print("거스름돈: 10원: {%d}개" %ten)

```
